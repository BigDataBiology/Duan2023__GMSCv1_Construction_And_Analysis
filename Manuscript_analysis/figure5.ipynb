```python
import seaborn as sns
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
plt.rcParams['svg.fonttype'] = 'none'
```

```python
sensitivity = pd.read_csv('./data/figure5/sensitivity.csv')
sensitivity
```

```python
diamond = sensitivity[sensitivity['Tool']=='Diamond']
diamond
```

```python
diamond
```

```python
colors= ['#1b9e77','#d95f02','#7570b3','#e7298a','#66a61e']
sns.catplot(x="Sensitivity", y="Rescue", hue="Length",kind="point",palette=sns.color_palette(colors),data=diamond)
plt.xticks(rotation=90)
plt.xlabel('')
plt.ylabel('Number of smORFs')
```

```python
mmseqs = sensitivity[sensitivity['Tool']=='MMseqs']
mmseqs
```

```python
colors= ['#1b9e77','#d95f02','#7570b3','#e7298a','#66a61e']
sns.catplot(
    data=mmseqs, x="Sensitivity", y="Rescue", hue="Length",
    kind="point",palette=sns.color_palette(colors)
)
plt.xlabel('')
plt.ylabel('Number of smORFs')
```

```python
time = sensitivity[sensitivity['Time(s)'].notna()]
time_diamond = time[time['Tool']=='Diamond']
time_mmseqs = time[time['Tool']=='MMseqs']
```

```python
sns.catplot(
    data=time_diamond, x="Sensitivity", y="Time(s)",
    kind="point",color='#7570b3'
)
plt.xticks(rotation=90)
plt.xlabel('')
plt.ylabel('Time (s)')
plt.tight_layout()
```

```python
sns.catplot(
    data=time_mmseqs, x="Sensitivity", y="Time(s)",
    kind="point",color='#a6761d'
)
plt.ylim(0,8000)
#plt.xticks(rotation=90)
plt.xlabel('')
plt.ylabel('Time (s)')
plt.tight_layout()
```
